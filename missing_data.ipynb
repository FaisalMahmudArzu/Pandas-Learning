```python
import numpy as np 
import pandas as pd 
```

## Finding Missing data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Searching the data if null or not
df.isna()
```

```python
# Checking how many null values are there
df.isna().sum()
```

```python
# Column wise null value checking
df.isna().any()
```

# Removing missing data

```python
df
```

```python
# To filter out non null row. Works row wise 
df.dropna()
```

```python
# Thresh means how many non null data should be there in one row
df.dropna(thresh=3)
```

# Filling the missing table

```python
df
```

```python
df.fillna(0)
```

```python
values = {'A':0,'B':100,'C':300,'D':400}
# column wise selection
df.fillna(value=values)
```

```python
# If I want to fill NAN values with mean
df.fillna(df.mean())
```
