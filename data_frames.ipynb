```python
import numpy as np
import pandas as pd
```

### Creating a Dataframe

```python
data = {
    'Name': ['John','Anna','Peter','Linda'],
    'Age': [28,34,29,42],
    'City': ['New York','Paris','Berlin','London'],
    'Salary': [65000,70000,62000,85000]
}
df = pd.DataFrame(data)
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
# Without setting columns
df2 = pd.DataFrame(data_list)
columns = ["Name","Age","City","Salary"]
# With setting columns
df2 = pd.DataFrame(data_list, columns=columns)
df2
```

### Selection and indexing of columns

```python
df2
```

```python
# Targeting columns
df2[['Name','City']]
```

### Creating a new column

```python
df2["Designation"] = ["Doctor","Engineer","Doctor","Engineer"]
```

```python
df2
```

### Removing Column

```python
# Use inplace if you want to remove column from memory too
df2.drop('Designation',axis = 1,inplace = True)
```

```python
df2
```

```python
#removing row
df2.drop(0,axis = 0)
```

### Conditional Selection

```python
df2
```

```python
# I only want to see those people whose age is above 30

df2[df2['Age'] > 30]
```

```python
# I only want to see those people whose age is above 30 and their city must be Paris
df2[(df2['Age'] > 30) & (df2['City'] == 'Paris')]
```
